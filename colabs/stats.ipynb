```python
import pandas as pd ## pip install pandas
```

```python
with open('../word_count/reduce.txt', 'r', encoding='utf-8') as file:
    key = []
    value = []
    for line in file:
        parts = line.split()
        if len(parts) == 2:
            word, count = parts[0], int(parts[1])
            # dict_words[word] = count
            key.append(word)
            value.append(count)
```

```python
dict_words = {
    "key": key,
    "value": value
}
```

```python
df = pd.DataFrame(dict_words)
df
```
