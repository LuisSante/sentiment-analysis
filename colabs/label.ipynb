```python
dataset = pd.read_csv('../out/yape_label_18_11.csv')
dataset
```

```python
dataset['content_join'] = dataset['content_join'].astype(str).fillna('')
dataset
```

```python
positives = []
negatives = []

with open('../out/positivo.txt', 'r') as positive_words_file:
    for line in positive_words_file:
        positives.append(line.strip())

with open('../out/negativo.txt', 'r') as negatives_words_file:
    for line in negatives_words_file:
        negatives.append(line.strip())
```

```python
last_target = []
positives_set = set(positives) 
negatives_set = set(negatives)

for sentence in dataset['content_join']:
    count_positives = 0
    count_negatives = 0

    sentence_tokens = sentence.lower().split()
    for word in positives:
        if word in sentence_tokens:
            count_positives += 1

    for word in negatives:
        if word in sentence_tokens:
            count_negatives += 1
    
    if count_positives > count_negatives:
        last_target.append('Positivo')
    
    elif count_positives < count_negatives:
        last_target.append('Negativo')
    else:
        last_target.append('Neutral')
```

```python
from collections import Counter

element_counts = Counter(last_target)
print(element_counts)
```

```python
dataset['last_target'] = last_target
```
